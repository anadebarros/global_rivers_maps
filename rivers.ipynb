```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
rivers_na = gpd.read_file('rivers/rivers_noram_37341')
rivers_sa = gpd.read_file('rivers/rivers_samerica_37330')
rivers_eu = gpd.read_file('rivers/rivers_europe_37253')
rivers_ca = gpd.read_file('rivers/rivers_centam_37249')
rivers_au = gpd.read_file('rivers/rivers_australia_37252')
rivers_as = gpd.read_file('rivers/rivers_asia_37331')

rivers = [rivers_na, rivers_sa, rivers_eu, rivers_ca, rivers_au, rivers_as]
river_names = ['north_america', 'south_america', 'europe', 'central_america', 'australia','asia']
```

```python
for i, river in enumerate(rivers):

    #converting the Strahler index into line-width values
    river['width'] = 10 - river.A_Strahler
    river['width'] = river['width']/float(5)

    #creating a canvas in matplotli
    f, ax = plt.subplots(1,1,figsize=(12,15))

    #doing the plotting
    river.plot(column = 'A_Strahler', ax = ax, linewidth = river['width'], cmap = 'Blues')

    #editing the plot
    f.patch.set_facecolor('k')
    ax.set_facecolor('k')
    ax.axis('off')

    #saving the visual
    plt.savefig(f'rivers_{river_names[i % len(river_names)]}.png', dpi = 200, bbox_inches = 'tight')
    plt.show()
```
